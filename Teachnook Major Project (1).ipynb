```python
#import dataset using pandans
import pandas as pd
data = pd.read_csv(r"C:\Users\Lenovo\Dropbox\PC\Desktop\Dataset.csv")
print(data)
```

```python
#Data Cleaning
#1. Remove the rows which has null values
data.dropna(inplace = True)
print(data)
```

```python
#2. Correct the format in the given data
data["YearBuilt"] = data["YearBuilt"].astype('int')
data['Date'] = pd.to_datetime(data['Date'])
print(data)
```

```python
#3. Given that there are no hoses having more than 6 rooms. So, removing rows which has more than 6 rooms:
for i in data.index:
    if data.loc[i, "Rooms"]>6:
        data.drop(i, inplace = True)
print(data)
```

```python
#5. Remving rows which has 0 as size of the land
data = data[data["Landsize"] > 0]
data
```

3. Logiistic Regression

```python
price_data = []
for i in data.index:
    if data.loc[i, "Price"]<=2000000:
        price_data.append("0 Lakh to 20 Lakh") 
    elif data.loc[i, "Price"]<=3000000:
        price_data.append("20 Lakh to 30 Lakh")
    elif data.loc[i, "Price"]<=4000000:
        price_data.append("30 Lakh to 40 Lakh")
    elif data.loc[i, "Price"]<=5000000:
        price_data.append("40 Lakh to 50 Lakh")
    elif data.loc[i, "Price"]<=6000000:
        price_data.append("50 Lakh to 60 Lakh")
    elif data.loc[i, "Price"]<=7000000:
        price_data.append("60 Lakh to 70 Lakh")
    elif data.loc[i, "Price"]<=8000000:
        price_data.append("70 Lakh to 80 Lakh")
    elif data.loc[i, "Price"]<=9000000:
        price_data.append("80 Lakh to 90 Lakh")
len(price_data)
```

```python
new_data = data.assign(Price_cluster = price_data)
new_data
```

```python
#Assigning x and y values
x = new_data[["Rooms", "Distance", "Landsize", "Bedroom2", "Bathroom", "Car", "Propertycount", "BuildingArea"]]
y = new_data["Price_cluster"]
```

```python
#Import Linear Regression
from sklearn.linear_model import  LogisticRegression
model = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_Y, test_Y = train_test_split(x, y)
model.fit(train_X, train_Y)
Y_pred = model.predict(test_X)
```

```python
Y_pred
```

```python
test_Y
```

```python
#To test the error value between predicted and actual value
from sklearn.metrics import accuracy_score
accuracy_score(Y_pred, test_Y)*100
```

4. Clustering

```python
X = new_data.iloc[:, [8, 4]]
X
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
y_kmeans = list(y_kmeans)
y_kmeans
X = X.assign(Y_kmeans = y_kmeans)
X
```

```python
a = X.loc[X['Y_kmeans'] == 0]
a
```

```python
b = a["Distance"].tolist()
print(b)
c = a["Price"].tolist()
print(c)
```

```python
d = X.loc[X['Y_kmeans'] == 1]
d
```

```python
e = d["Distance"].tolist()
print(e)
f = d["Price"].tolist()
print(f)
```

```python
g = X.loc[X['Y_kmeans'] == 2]
g
```

```python
h = g["Distance"].tolist()
print(h)
i = g["Price"].tolist()
print(i)
```

```python
plt.scatter(b, c, s = 0.5, c = 'red', label = 'Cluster 1')
plt.scatter(e, f, s = 0.5, c = 'blue', label = 'Cluster 2')
plt.scatter(h, i, s = 0.5, c = 'green', label = 'Cluster 3')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 100, c = 'yellow', label = 'Centroids')
plt.xlabel('Landsize')
plt.ylabel('Price')
plt.legend()
plt.show()
```

In the above clustering, we can conclude, if the distance decreases the price tends to increase in most of the cases
